```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
auto_data= pd.read_csv('auto_data.csv')
```

```python
auto_data.head()
```

```python
#function to define origin
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'Europe'
    else:
        return 'Africa'
#use apply function
auto_data['origin']= auto_data['origin'].apply(origin)
```

```python
auto_data.head()
```

```python
#draw the pairplot using sns for mpg, weight , originand with hue origin, set the size to 4
sns.pairplot(auto_data[['mpg','weight', 'origin']], hue='origin', height=4)
```
